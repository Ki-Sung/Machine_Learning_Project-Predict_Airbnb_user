### Age Gender buket 

```python
age_gender = pd.read_csv('./datas2/age_gender_bkts.csv')
age_gender
```

- Age Gender bucket (연령대 & 성별)
    - age_bucket : 연령대
    - country_destination : 목적지 (국가)
    - gender : 성별
    - population in thousands : 여행객수 (1000명)
    - year : 연도 

```python
age_gender['year'] = age_gender['year'].astype(int)
```

```python
age_gender['age_bucket'].value_counts()
```

```python
age_gender['age_bucket'] = age_gender['age_bucket'].replace("0-4", "1")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("5-9", "1")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("10-14", "10")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("15-19", "10")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("20-24", "20")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("25-29", "20")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("30-34", "30")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("35-39", "30")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("40-44", "40")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("45-49", "40")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("50-54", "50")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("55-59", "50")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("60-64", "60")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("65-69", "60")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("70-74", "70")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("75-79", "70")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("80-84", "80")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("85-89", "80")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("85-89", "80")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("90-94", "90")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("95-99", "90")
age_gender['age_bucket'] = age_gender['age_bucket'].replace("100+", "100")
age_gender['age_bucket']= age_gender['age_bucket'].astype(int)
```

```python
age_gender.info()
```

```python
gender = age_gender.groupby("gender").agg('sum')[['population_in_thousands']].round().reset_index()
gender.sort_values('population_in_thousands', ascending=False, inplace=True)
gender
```

```python
from pyecharts import Bar

keys = list(gender["gender"].values)
values = list(gender["population_in_thousands"].values)

bar = Bar("성별 여행객수")
bar.add("단위:1000", keys, values, is_stack=True, is_label_show=True, label_pos='inside', 
        label_color=['#0067A3'], legend_pos='right')

bar.height = 500
bar.width = 900

bar
```

- 여성 여행객이 많음.

```python
country = age_gender.groupby("country_destination").agg('sum')[['population_in_thousands']].round().reset_index()
country.sort_values('population_in_thousands', ascending=False, inplace=True)
country
```

```python
keys = list(country["country_destination"].values)
values = list(country["population_in_thousands"].values)

bar = Bar("목적지 국가 여행객수")
bar.add("단위:1000", keys, values, is_stack=True, is_label_show=True, label_pos='inside', 
        label_color=['#0067A3'], legend_pos='right')

bar.height = 500
bar.width = 1000

bar
```

- 목적지 국가별 여행객수는 미국 : 3억 2천만명, 독일 : 8천2백만명, 프랑스 : 6천4백만명, 영국 : 6천3백만명,  

```python
age = age_gender.groupby("age_bucket").agg('sum')[['population_in_thousands']].round().reset_index()
age['population_in_thousands']
age
```

```python
keys = list(age["age_bucket"].values)
values = list(age["population_in_thousands"].values)

bar = Bar("연령별 국가 여행객수")
bar.add("단위:1000", keys, values, is_stack=True, is_label_show=True, label_pos='inside', 
        label_color=['#0067A3'], legend_pos='right')

bar.height = 500
bar.width = 1000

bar
```
